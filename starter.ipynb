```python
import pandas as pd
import yfinance as yf
import datetime 
from datetime import date, timedelta
```

```python
today = date.today()
today
```

```python
d1 = today.strftime('%Y-%m-%d')
end_date = d1
d2 = date.today() - timedelta(days=720)
d2 = d2.strftime('%Y-%m-%d')
start_date = d2
```

```python
data = yf.download('AMC',
                    start=start_date,
                    end=end_date,
                    progress=False)
print(data.head())
```

```python
import plotly.express as px
figure = px.line(data, x = data.index,
                    y = "Close",
                    title = "TSA Line Plot")
figure.show()
```

```python
import plotly.graph_objects as go
figure = go.Figure(data=[go.Candlestick(x = data.index,
                                        open = data["Open"],
                                        high = data["High"],
                                        low = data["Low"],
                                        close = data["Close"])])
figure.update_layout(title = "TSA Candlestick",
                        xaxis_rangeslider_visible = False)
figure.show()
```

```python
figure = px.bar(data, x = data.index,
                y = "Close",
                title = "TSA Bar")
figure.show()
```

```python
figure = px.line(data, x = data.index,
                    y = 'Close',
                    range_x = ['2022-01-01','2022-03-29'],
                    title = "TSA Custom Date")
figure.show()
```

```python
figure = go.Figure(data = [go.Candlestick(x=data.index,
                                        open = data["Open"],
                                        high = data["High"],
                                        low = data["Low"],
                                        close = data["Close"])])
figure.update_layout(title="TSA Interactive Candlestick")

figure.update_xaxes(
    rangeslider_visible = True,
    rangeselector = dict(
        buttons = list([
            dict(count = 1, label = "1m", step = "month", stepmode = "backward"),
            dict(count = 6, label = "6m", step = "month", stepmode = "backward"),
            dict(count = 1, label = "YTD", step = "year", stepmode = "todate"),
            dict(count = 1, label = "1y", step = "year", stepmode = "backward"),
            dict(step = "all")
        ])
    )
)
figure.show()
```
